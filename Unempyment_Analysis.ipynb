```python
# üíº Unemployment Analysis Project

# 1Ô∏è‚É£ Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2Ô∏è‚É£ Load the Dataset
# üëâ If you have your own dataset, upload it in Colab and replace the filename below
# Example dataset name: "Unemployment_Rate.csv"
# df = pd.read_csv('/content/Unemployment_Rate.csv')

# For demo, let‚Äôs create a small sample unemployment dataset
data = {
    'Month': ['Jan-2020','Feb-2020','Mar-2020','Apr-2020','May-2020','Jun-2020','Jul-2020',
              'Aug-2020','Sep-2020','Oct-2020','Nov-2020','Dec-2020','Jan-2021','Feb-2021'],
    'Unemployment_Rate': [6.5, 6.2, 7.0, 13.5, 11.7, 10.2, 9.8, 8.4, 8.1, 7.9, 7.6, 7.4, 7.1, 6.8]
}

df = pd.DataFrame(data)
print("‚úÖ Dataset Loaded Successfully!")
print(df.head())

# 3Ô∏è‚É£ Convert 'Month' column to datetime format
df['Month'] = pd.to_datetime(df['Month'], format='%b-%Y')

# 4Ô∏è‚É£ Basic Data Information
print("\nüìã Dataset Info:")
print(df.info())

print("\nüìà Summary Statistics:")
print(df.describe())

# 5Ô∏è‚É£ Data Cleaning (check for missing values)
print("\nüîç Missing Values:")
print(df.isnull().sum())

# 6Ô∏è‚É£ Data Visualization

plt.figure(figsize=(10,6))
sns.lineplot(x='Month', y='Unemployment_Rate', data=df, marker='o', color='red')
plt.title('üìä Unemployment Rate Over Time')
plt.xlabel('Month')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()

# 7Ô∏è‚É£ Highlight COVID-19 Impact Period
plt.figure(figsize=(10,6))
sns.barplot(x='Month', y='Unemployment_Rate', data=df, palette='coolwarm')
plt.title('ü¶† Covid-19 Impact on Unemployment Rates (2020‚Äì2021)')
plt.xticks(rotation=45)
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('Month')
plt.show()

# 8Ô∏è‚É£ Identify Highest and Lowest Unemployment
highest = df[df['Unemployment_Rate'] == df['Unemployment_Rate'].max()]
lowest = df[df['Unemployment_Rate'] == df['Unemployment_Rate'].min()]

print("\nüìå Highest Unemployment:")
print(highest)

print("\nüìâ Lowest Unemployment:")
print(lowest)

# 9Ô∏è‚É£ Insights
print("\nüí° INSIGHTS:")
print("1. Unemployment spiked sharply in April 2020 due to COVID-19 lockdowns.")
print("2. The rate gradually declined from mid-2020 as restrictions eased.")
print("3. By early 2021, unemployment was close to pre-pandemic levels.")
print("4. Seasonal or recovery trends show improving job conditions after crisis.")
print("5. Policy support and reopening measures likely helped the decline.")
```
